### 1.3 ìˆ˜ì¹˜ ì˜ˆì œ

```python
# ìˆ˜ì¹˜ ì˜ˆì œ: ë²¡í„° ë‚´ì 
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 1.1: ë²¡í„° ë‚´ì ")
print("=" * 60)

# ë²¡í„° ì •ì˜
a = np.array([2, 3, 4])
b = np.array([1, 5, 2])

# ë‚´ì  ê³„ì‚° (ë‘ ê°€ì§€ ë°©ë²•)
dot_manual = sum(a[i] * b[i] for i in range(len(a)))
dot_numpy = np.dot(a, b)

print(f"ë²¡í„° a = {a}")
print(f"ë²¡í„° b = {b}")
print(f"\në‹¨ê³„ë³„ ê³„ì‚°:")
print(f"  aÂ·b = (2Ã—1) + (3Ã—5) + (4Ã—2)")
print(f"      = 2 + 15 + 8")
print(f"      = {dot_manual}")
print(f"\nNumPy ê²°ê³¼: {dot_numpy}")

# ê²€ì¦
assert dot_manual == dot_numpy, "ë‚´ì  ê³„ì‚° ì˜¤ë¥˜!"
print("\nâœ“ ê²€ì¦ ì™„ë£Œ: ìˆ˜ë™ ê³„ì‚°ê³¼ NumPy ê²°ê³¼ ì¼ì¹˜")

# ìˆ˜ì¹˜ ì˜ˆì œ: í–‰ë ¬ ê³±ì…ˆ
print("\n" + "=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 1.2: í–‰ë ¬ ê³±ì…ˆ")
print("=" * 60)

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

C = np.dot(A, B)

print(f"í–‰ë ¬ A (2Ã—2):\n{A}")
print(f"\ní–‰ë ¬ B (2Ã—2):\n{B}")
print(f"\nê²°ê³¼ C = A Ã— B:\n{C}")
print(f"\në‹¨ê³„ë³„ ê³„ì‚° (C[0,0]):")
print(f"  C[0,0] = A[0,:]Â·B[:,0] = (1Ã—5) + (2Ã—7) = 5 + 14 = 19")

# ê²€ì¦
assert C[0, 0] == 19, "í–‰ë ¬ ê³±ì…ˆ ê³„ì‚° ì˜¤ë¥˜!"
print("\nâœ“ ê²€ì¦ ì™„ë£Œ")
```

### 1.4 ì „ì¹˜, ì—­í–‰ë ¬, ê³ ìœ ê°’

```python
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 1.3: ì „ì¹˜, ì—­í–‰ë ¬, ê³ ìœ ê°’")
print("=" * 60)

# ì „ì¹˜ (Transpose)
A = np.array([[1, 2, 3],
              [4, 5, 6]])
A_T = A.T

print(f"ì›ë³¸ í–‰ë ¬ A (2Ã—3):\n{A}")
print(f"\nì „ì¹˜ A^T (3Ã—2):\n{A_T}")

# ì—­í–‰ë ¬ (Inverse)
B = np.array([[4, 7],
              [2, 6]])
B_inv = np.linalg.inv(B)
identity = np.dot(B, B_inv)

print(f"\ní–‰ë ¬ B:\n{B}")
print(f"\nì—­í–‰ë ¬ B^(-1):\n{B_inv}")
print(f"\nB Ã— B^(-1) (ë‹¨ìœ„í–‰ë ¬ì´ì–´ì•¼ í•¨):\n{identity}")

# ê³ ìœ ê°’ê³¼ ê³ ìœ ë²¡í„° (Eigenvalues and Eigenvectors)
eigenvalues, eigenvectors = np.linalg.eig(B)

print(f"\nê³ ìœ ê°’ (eigenvalues): {eigenvalues}")
print(f"ê³ ìœ ë²¡í„° (eigenvectors):\n{eigenvectors}")

# ê²€ì¦: AÂ·v = Î»Â·v
lambda_1 = eigenvalues[0]
v_1 = eigenvectors[:, 0]
left = np.dot(B, v_1)
right = lambda_1 * v_1

print(f"\nê²€ì¦ (BÂ·vâ‚ = Î»â‚Â·vâ‚):")
print(f"  BÂ·vâ‚ = {left}")
print(f"  Î»â‚Â·vâ‚ = {right}")
assert np.allclose(left, right), "ê³ ìœ ê°’/ê³ ìœ ë²¡í„° ê²€ì¦ ì‹¤íŒ¨!"
print("\nâœ“ ê³ ìœ ê°’/ê³ ìœ ë²¡í„° ê´€ê³„ ê²€ì¦ ì™„ë£Œ")
```

## 2. ë¯¸ë¶„ê³¼ í¸ë¯¸ë¶„ ğŸŸ¢

### 2.1 ì´ë¡  ì„¤ëª…

ì‹ ê²½ë§ í•™ìŠµì€ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ìµœì†Œí™”í•˜ëŠ” ê³¼ì •ì´ë©°, ì´ë¥¼ ìœ„í•´ ë¯¸ë¶„(ê¸°ìš¸ê¸°)ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

### 2.2 í•µì‹¬ ê°œë…

#### ì¼ë³€ìˆ˜ í•¨ìˆ˜ì˜ ë¯¸ë¶„

**ìˆ˜ì‹:**
$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

#### í¸ë¯¸ë¶„ (Partial Derivative)

**ìˆ˜ì‹:**
$$\frac{\partial f}{\partial x_i} = \lim_{h \to 0} \frac{f(x_1, \ldots, x_i + h, \ldots, x_n) - f(x_1, \ldots, x_i, \ldots, x_n)}{h}$$

#### ê¸°ìš¸ê¸° ë²¡í„° (Gradient)

**ìˆ˜ì‹:**
$$\nabla f = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \\ \vdots \\ \frac{\partial f}{\partial x_n} \end{bmatrix}$$

#### ê¸°í˜¸ ì„¤ëª…

| ê¸°í˜¸ | ì˜ë¯¸ |
|------|------|
| $f'(x)$ | fì˜ xì— ëŒ€í•œ ë¯¸ë¶„ |
| $\frac{\partial f}{\partial x_i}$ | fì˜ $x_i$ì— ëŒ€í•œ í¸ë¯¸ë¶„ |
| $\nabla f$ | ê¸°ìš¸ê¸° ë²¡í„° (gradient) |
| $h$ | ì•„ì£¼ ì‘ì€ ë³€í™”ëŸ‰ |

```python
# ìˆ˜ì¹˜ ì˜ˆì œ: ìˆ˜ì¹˜ ë¯¸ë¶„
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 2.1: ìˆ˜ì¹˜ ë¯¸ë¶„")
print("=" * 60)

def f(x):
    """í•¨ìˆ˜: f(x) = x^2"""
    return x ** 2

def numerical_derivative(func, x, h=1e-5):
    """ìˆ˜ì¹˜ ë¯¸ë¶„"""
    return (func(x + h) - func(x - h)) / (2 * h)

def analytical_derivative(x):
    """í•´ì„ì  ë¯¸ë¶„: f'(x) = 2x"""
    return 2 * x

x = 3.0
num_deriv = numerical_derivative(f, x)
ana_deriv = analytical_derivative(x)

print(f"í•¨ìˆ˜: f(x) = xÂ²")
print(f"ì : x = {x}")
print(f"\nìˆ˜ì¹˜ ë¯¸ë¶„: f'({x}) â‰ˆ {num_deriv:.6f}")
print(f"í•´ì„ì  ë¯¸ë¶„: f'({x}) = {ana_deriv:.6f}")
print(f"ì˜¤ì°¨: {abs(num_deriv - ana_deriv):.10f}")

assert np.isclose(num_deriv, ana_deriv, atol=1e-4), "ë¯¸ë¶„ ê³„ì‚° ì˜¤ë¥˜!"
print("\nâœ“ ê²€ì¦ ì™„ë£Œ: ìˆ˜ì¹˜ ë¯¸ë¶„ê³¼ í•´ì„ì  ë¯¸ë¶„ ì¼ì¹˜")
```

```python
# ìˆ˜ì¹˜ ì˜ˆì œ: í¸ë¯¸ë¶„ê³¼ ê¸°ìš¸ê¸°
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 2.2: í¸ë¯¸ë¶„ê³¼ ê¸°ìš¸ê¸°")
print("=" * 60)

def g(x, y):
    """í•¨ìˆ˜: g(x,y) = xÂ² + 2xy + yÂ²"""
    return x**2 + 2*x*y + y**2

def gradient_g(x, y):
    """í•´ì„ì  ê¸°ìš¸ê¸°: âˆ‡g = [2x + 2y, 2x + 2y]"""
    return np.array([2*x + 2*y, 2*x + 2*y])

def numerical_gradient(func, x, y, h=1e-5):
    """ìˆ˜ì¹˜ì  ê¸°ìš¸ê¸°"""
    grad_x = (func(x + h, y) - func(x - h, y)) / (2 * h)
    grad_y = (func(x, y + h) - func(x, y - h)) / (2 * h)
    return np.array([grad_x, grad_y])

x, y = 1.0, 2.0
num_grad = numerical_gradient(g, x, y)
ana_grad = gradient_g(x, y)

print(f"í•¨ìˆ˜: g(x,y) = xÂ² + 2xy + yÂ²")
print(f"ì : (x, y) = ({x}, {y})")
print(f"\nìˆ˜ì¹˜ì  ê¸°ìš¸ê¸°: {num_grad}")
print(f"í•´ì„ì  ê¸°ìš¸ê¸°: {ana_grad}")
print(f"\në‹¨ê³„ë³„ ê³„ì‚°:")
print(f"  âˆ‚g/âˆ‚x = 2x + 2y = 2({x}) + 2({y}) = {ana_grad[0]}")
print(f"  âˆ‚g/âˆ‚y = 2x + 2y = 2({x}) + 2({y}) = {ana_grad[1]}")

assert np.allclose(num_grad, ana_grad, atol=1e-4), "ê¸°ìš¸ê¸° ê³„ì‚° ì˜¤ë¥˜!"
print("\nâœ“ ê²€ì¦ ì™„ë£Œ")
```

### 2.3 ì•¼ì½”ë¹„ì•ˆê³¼ í—¤ì‹œì•ˆ

#### ì•¼ì½”ë¹„ì•ˆ í–‰ë ¬ (Jacobian Matrix)

ë²¡í„° í•¨ìˆ˜ $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$ì— ëŒ€í•´:

$$J = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{bmatrix}$$

#### í—¤ì‹œì•ˆ í–‰ë ¬ (Hessian Matrix)

ìŠ¤ì¹¼ë¼ í•¨ìˆ˜ $f: \mathbb{R}^n \to \mathbb{R}$ì— ëŒ€í•´:

$$H = \begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}$$

```python
# ìˆ˜ì¹˜ ì˜ˆì œ: ì•¼ì½”ë¹„ì•ˆ
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 2.3: ì•¼ì½”ë¹„ì•ˆ í–‰ë ¬")
print("=" * 60)

def vector_function(x):
    """ë²¡í„° í•¨ìˆ˜: f(x,y) = [xÂ²+y, xy]"""
    return np.array([x[0]**2 + x[1], x[0]*x[1]])

def jacobian_analytical(x):
    """í•´ì„ì  ì•¼ì½”ë¹„ì•ˆ"""
    return np.array([[2*x[0], 1],
                     [x[1], x[0]]])

x = np.array([2.0, 3.0])
J = jacobian_analytical(x)

print(f"ë²¡í„° í•¨ìˆ˜: f(x,y) = [xÂ²+y, xy]")
print(f"ì : x = {x}")
print(f"\nì•¼ì½”ë¹„ì•ˆ í–‰ë ¬ J:")
print(J)
print(f"\në‹¨ê³„ë³„ ê³„ì‚°:")
print(f"  âˆ‚fâ‚/âˆ‚x = 2x = 2({x[0]}) = {J[0,0]}")
print(f"  âˆ‚fâ‚/âˆ‚y = 1 = {J[0,1]}")
print(f"  âˆ‚fâ‚‚/âˆ‚x = y = {x[1]} = {J[1,0]}")
print(f"  âˆ‚fâ‚‚/âˆ‚y = x = {x[0]} = {J[1,1]}")

print("\nâœ“ ì•¼ì½”ë¹„ì•ˆ ê³„ì‚° ì™„ë£Œ")
```

## 3. ì—°ì‡„ë²•ì¹™ê³¼ ì—­ì „íŒŒ ğŸŸ¡

### 3.1 ì´ë¡  ì„¤ëª…

**ì—­ì „íŒŒ(Backpropagation)**ëŠ” ì‹ ê²½ë§ í•™ìŠµì˜ í•µì‹¬ ì•Œê³ ë¦¬ì¦˜ì…ë‹ˆë‹¤. ì—°ì‡„ë²•ì¹™(Chain Rule)ì„ ì‚¬ìš©í•˜ì—¬ ì†ì‹¤ í•¨ìˆ˜ì˜ ê·¸ë˜ë””ì–¸íŠ¸ë¥¼ íš¨ìœ¨ì ìœ¼ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤.

### 3.2 ì—°ì‡„ë²•ì¹™ (Chain Rule)

#### ì¼ë³€ìˆ˜ í•©ì„±í•¨ìˆ˜

$$y = f(g(x)) \Rightarrow \frac{dy}{dx} = \frac{dy}{dg} \cdot \frac{dg}{dx}$$

#### ë‹¤ë³€ìˆ˜ í•¨ìˆ˜

$$\frac{\partial L}{\partial x} = \sum_i \frac{\partial L}{\partial z_i} \cdot \frac{\partial z_i}{\partial x}$$

#### ê¸°í˜¸ ì„¤ëª…

| ê¸°í˜¸ | ì˜ë¯¸ |
|------|------|
| $L$ | ì†ì‹¤ í•¨ìˆ˜ (Loss function) |
| $z_i$ | ì¤‘ê°„ ë³€ìˆ˜ (intermediate variable) |
| $\frac{\partial L}{\partial x}$ | Lì˜ xì— ëŒ€í•œ í¸ë¯¸ë¶„ |

### 3.3 ë‹¨ì¼ ì€ë‹‰ì¸µ ì‹ ê²½ë§ì˜ ì—­ì „íŒŒ

```python
# ì™„ì „í•œ ì—­ì „íŒŒ êµ¬í˜„ (ë‹¨ì¼ ì€ë‹‰ì¸µ)
print("=" * 60)
print("ìˆ˜ì¹˜ ì˜ˆì œ 3.1: ë‹¨ì¼ ì€ë‹‰ì¸µ ì—­ì „íŒŒ")
print("=" * 60)

# í™œì„±í™” í•¨ìˆ˜ì™€ ë¯¸ë¶„
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

# ë„¤íŠ¸ì›Œí¬ êµ¬ì¡°: 2 ì…ë ¥ â†’ 3 ì€ë‹‰ â†’ 1 ì¶œë ¥
print("ë„¤íŠ¸ì›Œí¬ êµ¬ì¡°: 2 ì…ë ¥ â†’ 3 ì€ë‹‰(ReLU) â†’ 1 ì¶œë ¥(Sigmoid)")
print()

# ì…ë ¥ê³¼ íŒŒë¼ë¯¸í„°
X = np.array([[1.0], [0.5]])  # (2, 1)
y_true = np.array([[1.0]])     # (1, 1)

# ê°€ì¤‘ì¹˜ì™€ í¸í–¥ ì´ˆê¸°í™”
np.random.seed(42)
W1 = np.random.randn(3, 2) * 0.1  # (3, 2)
b1 = np.zeros((3, 1))              # (3, 1)
W2 = np.random.randn(1, 3) * 0.1  # (1, 3)
b2 = np.zeros((1, 1))              # (1, 1)

print("ì´ˆê¸° íŒŒë¼ë¯¸í„°:")
print(f"  W1 shape: {W1.shape}")
print(f"  W2 shape: {W2.shape}")
print()

# ìˆœë°©í–¥ ì „íŒŒ
print("=== ìˆœë°©í–¥ ì „íŒŒ ===")
Z1 = np.dot(W1, X) + b1
A1 = relu(Z1)
print(f"ì€ë‹‰ì¸µ Z1:\n{Z1.flatten()}")
print(f"ì€ë‹‰ì¸µ A1 (after ReLU):\n{A1.flatten()}")
print()

Z2 = np.dot(W2, A1) + b2
A2 = sigmoid(Z2)
print(f"ì¶œë ¥ì¸µ Z2: {Z2[0,0]:.6f}")
print(f"ì¶œë ¥ì¸µ A2 (ì˜ˆì¸¡): {A2[0,0]:.6f}")
print(f"ì‹¤ì œ ê°’: {y_true[0,0]}")
print()

# ì†ì‹¤ ê³„ì‚° (Binary Cross-Entropy)
epsilon = 1e-10
loss = -np.mean(y_true * np.log(A2 + epsilon) + (1 - y_true) * np.log(1 - A2 + epsilon))
print(f"ì†ì‹¤ (Cross-Entropy): {loss:.6f}")
print()

# ì—­ì „íŒŒ
print("=== ì—­ì „íŒŒ ===")

# ì¶œë ¥ì¸µ ê·¸ë˜ë””ì–¸íŠ¸
dZ2 = A2 - y_true  # Sigmoid + Cross-Entropyì˜ ê°„ë‹¨í•œ í˜•íƒœ
dW2 = np.dot(dZ2, A1.T)
db2 = dZ2

print(f"ì¶œë ¥ì¸µ ê·¸ë˜ë””ì–¸íŠ¸:")
print(f"  dZ2 = {dZ2[0,0]:.6f}")
print(f"  dW2 = {dW2.flatten()}")
print(f"  db2 = {db2[0,0]:.6f}")
print()

# ì€ë‹‰ì¸µ ê·¸ë˜ë””ì–¸íŠ¸
dA1 = np.dot(W2.T, dZ2)
dZ1 = dA1 * relu_derivative(Z1)
dW1 = np.dot(dZ1, X.T)
db1 = dZ1

print(f"ì€ë‹‰ì¸µ ê·¸ë˜ë””ì–¸íŠ¸:")
print(f"  dZ1 = {dZ1.flatten()}")
print(f"  dW1 shape = {dW1.shape}")
print()

# ê·¸ë˜ë””ì–¸íŠ¸ ê²€ì¦ (ìˆ˜ì¹˜ ë¯¸ë¶„)
print("=== ê·¸ë˜ë””ì–¸íŠ¸ ê²€ì¦ ===")

def compute_loss(W1, b1, W2, b2, X, y_true):
    """ì†ì‹¤ ê³„ì‚° í•¨ìˆ˜"""
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    loss = -np.mean(y_true * np.log(A2 + epsilon) + (1 - y_true) * np.log(1 - A2 + epsilon))
    return loss

# W2ì˜ í•œ ì›ì†Œì— ëŒ€í•´ ìˆ˜ì¹˜ ë¯¸ë¶„
h = 1e-5
W2_test = W2.copy()
W2_test[0, 0] += h
loss_plus = compute_loss(W1, b1, W2_test, b2, X, y_true)

W2_test = W2.copy()
W2_test[0, 0] -= h
loss_minus = compute_loss(W1, b1, W2_test, b2, X, y_true)

numerical_grad = (loss_plus - loss_minus) / (2 * h)
analytical_grad = dW2[0, 0]

print(f"W2[0,0]ì— ëŒ€í•œ ê·¸ë˜ë””ì–¸íŠ¸:")
print(f"  ìˆ˜ì¹˜ ë¯¸ë¶„: {numerical_grad:.8f}")
print(f"  ì—­ì „íŒŒ:    {analytical_grad:.8f}")
print(f"  ì°¨ì´:      {abs(numerical_grad - analytical_grad):.10f}")

assert np.isclose(numerical_grad, analytical_grad, atol=1e-5), "ê·¸ë˜ë””ì–¸íŠ¸ ê²€ì¦ ì‹¤íŒ¨!"
print("\nâœ“ ê·¸ë˜ë””ì–¸íŠ¸ ê²€ì¦ ì™„ë£Œ")
```
